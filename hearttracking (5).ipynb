# preprocessing

```python
sns.scatterplot(x=df['age'],y=df['chol'])
```

```python

sns.scatterplot(df['chol'])
```

```python
def outlierfree(dataCol):
    sorted(dataCol)
    Q1,Q3=np.percentile(dataCol,[25,75])
    IQR=Q3-Q1
    lower=Q1-(1.5*IQR)
    upper=Q3+(1.5*IQR)
    return lower,upper
    
```

```python
lower_chol,upper_chol=outlierfree(df['chol'])
#df['chol'].apply(outlierfree)
```

```python
lower_chol
```

```python
upper_chol
```

```python
lower_trtbps,upper_trtbps=outlierfree(df['trtbps'])
lower_trtbps,upper_trtbps
```

```python
lower_thalachh,upper_thalachh=outlierfree(df['thalachh'])
lower_thalachh,upper_thalachh
```

```python
lower_oldpeak,upper_oldpeak=outlierfree(df['oldpeak'])
lower_oldpeak,upper_oldpeak
```

```python
sorted(df['chol'],reverse=True)
```

```python
df['chol'].replace(list(df[df['chol']>upper_chol].chol),upper_chol,inplace=True)
#df['chol']=df['chol'].replace(df[df['chol']>upper_chol],upper_chol,inplace=True) 
sorted(df['chol'],reverse=True)
```

```python
sns.scatterplot(df['chol'])
```

```python
df['thalachh'].replace(list(df[df['thalachh']<lower_thalachh].thalachh),lower_thalachh,inplace=True)
sorted(df['thalachh'],reverse=True)
```

```python
df['trtbps'].replace(list(df[df['trtbps']>upper_trtbps].trtbps),upper_trtbps,inplace=True)
sorted(df['trtbps'],reverse=True)
```

```python
df['oldpeak'].replace(list(df[df['oldpeak']>upper_oldpeak].oldpeak),upper_oldpeak,inplace=True)
sorted(df['oldpeak'],reverse=True)
```
